```python
import altair as alt
import pandas as pd
```

```python
start_df = pd.read_csv("../New-Venture-s-Compass/dataset/data.csv")


start_df=start_df.drop(["Series Code"],axis=1)
start_df.rename(columns={'2014 [YR2014]':'2014',
       '2015 [YR2015]':'2015', '2016 [YR2016]':'2016', '2017 [YR2017]':'2017', '2018 [YR2018]':'2018',
       '2019 [YR2019]':'2019'},inplace=True)
start_df.head()
```

```python
emptycountries=start_df[(start_df["Series Name"] == "Cost of business start-up procedures (% of GNI per capita)") & (start_df["2014"]=='..')
           & (start_df["2015"]=='..') & (start_df["2016"]=='..') & (start_df["2017"]=='..') & (start_df["2018"]=='..')
           & (start_df["2019"]=='..')]['Country Name'].unique()
```

```python
emptycountries.tolist()
```

```python
newdf=pd.DataFrame()
newarr=[]
    
for index,row  in start_df.iterrows(): 
    if row['Country Name'] not in emptycountries:
        newarr.append(row) 

newdf=pd.DataFrame(newarr)
newdf.count()
```
